```python
df_cleaned['Prefixe de section'] = df_cleaned["Prefixe de section"].fillna(0)
df_cleaned['Reference cadastrale'] = (
    df_cleaned['Code departement'].apply(lambda x: str(x).zfill(3)) + " " +
    df_cleaned['Code commune'].apply(lambda x: str(x).zfill(3)) + " " + 
    df_cleaned['Prefixe de section'].apply(lambda x: str(int(x)).zfill(3)) + " " +
    df_cleaned['Section'].apply(lambda x : str(x).ljust(2, "0")) + " " 
)
```

```python
df_cleaned = df.dropna(axis=1, thresh=2000000)
```

```python
df_cleaned = df_cleaned.drop_duplicates()
```

```python
df_cleaned = df_cleaned.fillna(0)
```

```python
df_cleaned["Date mutation"] = pd.to_datetime(df_cleaned["Date mutation"],format = "%d/%m/%Y")
```

# Réalisation ACP

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64', 'int32', 'int16', 'int8']).columns
```

```python
numeric_cols
```

```python
df_numeric = df_cleaned[numeric_cols]
```

```python
df_numeric.head()
```

```python
scaler = StandardScaler()
df_standardized = scaler.fit_transform(df_numeric)
```

```python
df_standardized
```

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df_standardized)
```

```python
pca_components
```

```python
explained_variance = pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(10, 7))
```

```python
plt.subplot(1, 2, 1)
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Composantes Principales')
plt.ylabel('Ratio de Variance Expliquée')
```

```python
plt.subplot(1, 2, 2)
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], alpha=0.5)
plt.title('Projection des données sur les deux premières composantes principales')
plt.xlabel('Composante Principale 1')
plt.ylabel('Composante Principale 2')
```

```python
pca_df = pd.DataFrame(pca.components_, columns=numeric_cols, index=[f'PC{i+1}' for i in range(len(explained_variance))])
print(pca_df.head())
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(pca_df.iloc[:2].T, annot=True, cmap='coolwarm')
plt.title('Coefficients des variables pour les deux premières composantes principales')
plt.show()
```
