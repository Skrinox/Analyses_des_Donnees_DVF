# 1. Data preprocessing

Load the text file

```python
file_path = "valeursfoncieres-2023.txt"

df = pd.read_csv(file_path, delimiter='|')
```

Display the 5 first lines of the pandas DataFrame

```python
df.head()
```

```python
df.shape
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df_cleaned = df
```

```python
df_cleaned.head()
```

```python
df_cleaned['Prefixe de section'] = df_cleaned["Prefixe de section"].fillna(0)
df_cleaned['Reference cadastrale'] = (
    df_cleaned['Code departement'].apply(lambda x: str(x).zfill(3)) + " " +
    df_cleaned['Code commune'].apply(lambda x: str(x).zfill(3)) + " " + 
    df_cleaned['Prefixe de section'].apply(lambda x: str(int(x)).zfill(3)) + " " +
    df_cleaned['Section'].apply(lambda x : str(x).ljust(2, "0")) + " " 
)
```

```python
df_cleaned = df.dropna(axis=1, thresh=2000000)
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.drop_duplicates()
```

```python
df_cleaned = df_cleaned.fillna(0)
```

```python
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned["Date mutation"] = pd.to_datetime(df_cleaned["Date mutation"])
```

```python
df_cleaned.dtypes
```

```python
# Numerise data
```

```python
df_cleaned['Valeur fonciere'] = df_cleaned['Valeur fonciere'].replace(',', '.', regex=True)
```

```python
df_cleaned['Valeur fonciere'] = df_cleaned['Valeur fonciere'].astype(float)
```

```python
df_cleaned.loc[df_cleaned['Code departement'] == '2A', 'Code departement'] = 200
df_cleaned.loc[df_cleaned['Code departement'] == '2B', 'Code departement'] = 201
```

```python
df_cleaned[df_cleaned['Code departement'] == 200]
```

```python
df_cleaned['Code departement'] = df_cleaned['Code departement'].astype(int)
```

```python
df_cleaned['Reference cadastrale'] = df_cleaned['Reference cadastrale'].astype()
```

```python
df_cleaned.dtypes
```

```python
# Encoding 
```

```python
df_cleaned['Nature mutation code'] = df_cleaned["Nature mutation"].astype("category").cat.codes
```

```python
df_cleaned['Type de voie code'] = df_cleaned["Type de voie"].astype("category").cat.codes
```

```python
df_cleaned.dtypes
```

```python
df_cleaned['Section code'] = df_cleaned["Section"].astype("category").cat.codes
```

```python
df_cleaned["Nature culture code"] = df_cleaned["Nature culture"].astype("category").cat.codes
```
